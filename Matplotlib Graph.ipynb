# My Graph

```python
#Import modules

import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np
```

```python
#We'll just make an alright, half-way decent line graph of our data.

#Plotting out the line graphs.

HighIncomePlot, = plt.plot(years, high_income_nation_populations, 'g', label='High Income Pop', color = '#85bb65')
LowIncomePlot, = plt.plot(years,  low_income_nation_populations, 'g', label='Low Income Pop', color = '#FF2323')
WorldPopPlot, = plt.plot(years,  world_population, 'g', label='World Pop', color = '#5b92e5')

#Legend.
plt.legend(loc='upper left')

#Specifically there's 56 years for this range so... four years per line makes it look nice and neat.

plt.xticks(years[::4])

fig = plt.gcf()

#Big enough for our needs.
fig.set_size_inches(15, 10)
#Grid for comparison.
plt.grid(True)
#Title so you know what it's about. Everything is sized to taste.
plt.title('High Income and Low Income Populations, compared to the World, size = 15)
#Labels, sized to taste also.
plt.xlabel('Year', size = 15)
plt.ylabel('Population in Billions', size = 15)
#Bingo.
plt.show()
```

The above graph lets us make use of all of our data sets, cleanly shows in an examinable fashion how the divide between high income nad low income populations is shrinking, while still making it apparent there are many people who fall into neither a high or low income category. It also has titles, labels, and a legend. The colors are specifically "money green" for high income, "danger red" for low income, and "United Nations Blue" for the world population, to invoke an idea of how those trends are actually going along.
